```python
import pandas as pd

df = pd.read_excel('product_data9.xlsx')

min_values = []
max_values = []
left_values = []
prices = []
average_prices = []

for index, row in df.iterrows():
    min_val = None
    max_val = None
    left_val = None
    price_sum = 0
    price_values = []  # Список для хранения цен в текущей строке
    
    for i in range(2, len(row), 2):
        try:
            cell_value = float(row[i].replace('\xa0₽', '').replace(' ', ''))
            left_cell_value = row[i - 1]
            
            if min_val is None or cell_value < min_val:
                # Если минимальное значение еще не установлено или текущее значение меньше, обновляем минимальное значение и значение слева
                min_val = cell_value
                left_val = left_cell_value
            
            if max_val is None or cell_value > max_val:
                # Если максимальное значение еще не установлено или текущее значение больше, обновляем максимальное значение
                max_val = cell_value
            
            price_sum += cell_value  # Добавляем цену к сумме
            price_values.append(cell_value)  # Добавляем цену в список
        except (ValueError, AttributeError):
            continue
    
    # Записываем минимальное, максимальное значение и значение слева в соответствующие списки
    min_values.append(int(min_val))
    max_values.append(max_val)
    left_values.append(left_val)
    
    # Вычисляем среднее значение цен и добавляем его в список
    num_prices = len(price_values)  # Количество цен
    average_price = sum(price_values) / num_prices if num_prices > 0 else None
    average_prices.append(average_price)

# Создаем новые столбцы в датафрейме и записываем в них списки значений
df['Мин_цена'] = min_values
df['Средняя цена'] = average_prices
df['Макс_цена'] = max_values
df['Продавец Мин_цена'] = left_values

df['Разница мин_сред'] = df['Средняя цена'] - df['Мин_цена']
df['Разница мин_макс'] = df['Макс_цена'] - df['Мин_цена']

df['Разница мин_сред проц'] = (df['Средняя цена'] - df['Мин_цена'])/df['Средняя цена']*100
df['Разница мин_макс проц'] = (df['Макс_цена'] - df['Мин_цена'])/df['Макс_цена']*100

df = df.sort_values(by='Разница мин_макс проц', ascending=False)

df.to_excel('product_data-9.xlsx', index=False)
```
